in this project i will use evolution technique to do gradient descent for a linear regression model


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

This Below code will generate the sample noisy linear data and store it into a vector called Y

```python
x=list(range(20))
Y=[]
for xi in x :
  Y.append(2.6*xi+1.8+(random.randint(-285,286))/70)
Y=np.array(Y)
x=np.array(x)
print(Y)
plt.scatter(x,Y)
plt.show()
```

```python
def MSE(a,b):
  global x,Y
  y=a*x+b
  D=Y-y
  E=np.dot(D,D)
  return (E/20)**0.5


```

```python
#7th Roots of unity points in 2D Plane
r1=np.array([np.cos((2*np.pi*(0))/7),np.sin((2*np.pi*(0))/7)])
r2=np.array([np.cos((2*np.pi*(1))/7),np.sin((2*np.pi*(1))/7)])
r3=np.array([np.cos((2*np.pi*(2))/7),np.sin((2*np.pi*(2))/7)])
r4=np.array([np.cos((2*np.pi*(3))/7),np.sin((2*np.pi*(3))/7)])
r5=np.array([np.cos((2*np.pi*(4))/7),np.sin((2*np.pi*(4))/7)])
r6=np.array([np.cos((2*np.pi*(5))/7),np.sin((2*np.pi*(5))/7)])
r7=np.array([np.cos((2*np.pi*(6))/7),np.sin((2*np.pi*(6))/7)])
```

```python
Lead_point=np.array([random.random(),random.random()])
Learning_rate=float(input('Learning Rate : '))
for i in range(10000):
    d={MSE(*(Lead_point+(Learning_rate*r1))):(Lead_point+(Learning_rate*r1)),
       MSE(*(Lead_point+(Learning_rate*r2))):(Lead_point+(Learning_rate*r2)),
       MSE(*(Lead_point+(Learning_rate*r3))):(Lead_point+(Learning_rate*r3)),
       MSE(*(Lead_point+(Learning_rate*r4))):(Lead_point+(Learning_rate*r4)),
       MSE(*(Lead_point+(Learning_rate*r5))):(Lead_point+(Learning_rate*r5)),
       MSE(*(Lead_point+(Learning_rate*r6))):(Lead_point+(Learning_rate*r6)),
       MSE(*(Lead_point+(Learning_rate*r7))):(Lead_point+(Learning_rate*r7))}
    print("Loss : " ,min(d.keys()))
    Lead_point=d[min(d.keys())]
a,b=(Lead_point)
```

```python
print(a,b)
plt.scatter(x,Y)

p=np.linspace(0,20,100)
q=p*a+b

plt.plot(p,q)
plt.show()
```
